```python
#1) 
import plotly.express as px
import seaborn as sns
```

```python
titanicds=sns.load_dataset('titanic')
titanicds
```

```python
px.scatter(x='fare',y='age',data_frame=titanicds)
```

```python
#2)
tipsds=sns.load_dataset('tips')
tipsds
```

```python
px.box(x='total_bill',y='tip',data_frame=tipsds)
```

```python
#3)
px.histogram(x="sex",y="total_bill",data_frame=tipsds,pattern_shape="smoker",color='day')
```

```python
#4)
irisds=sns.load_dataset('iris')
colums=["sepal_length", "sepal_width", "petal_length", "petal_width"]
px.scatter_matrix(dimensions=colums,color="species",data_frame=irisds)
```

```python
#5) A Distplot or distribution plot, depicts the variation in the data distribution. 
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(200)-2
x2 = np.random.randn(200)
x3 = np.random.randn(200)+2
x4 = np.random.randn(200)+4

data = [x1, x2, x3, x4]

labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

fig = ff.create_distplot(data, labels)
fig.show()
```
