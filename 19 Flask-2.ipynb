1) Get - Get method is used to send data over http without any encoding. There is restriction on sending data, maximum 2048 characters can be send at one time. Only ASCII characters are allowed. Data is visible to everyone in the url

Post - Post is used to send data by encoding it. There is no restriction on characters size. We can send binary data as well. Data is not visible to anyone as its encoded.

2) Request object contains all the data being send from client to server. This data can be retrieved using GET/POST methods.


 3) redirect() is used to redirect from current url to another url mentioned in the method. It opens the url mentioned in the method from current url.
 

4) Templates are files containing static data and placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates.
