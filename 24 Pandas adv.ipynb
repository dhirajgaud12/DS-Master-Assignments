```python
#3)
def calcsum(newdf):
    newdf.rename(columns = {0:'Values'}, inplace = True)
    tot=0
    for i in newdf['Values'][:3]:
        tot=tot+i
    return tot

calcsum(pd.DataFrame(pd.Series([10, 20, 30, 40, 50])))
    
```

```python
#4) 
def getWC(newdf,wordlist):
    newdf['Word_Count']=wordlist
    return newdf
    
    
thisdf=pd.DataFrame(pd.Series([1,2,3,4]))

wordlist=['aa','bbb','cccc','ddddd']
getWC(thisdf,wordlist)
```

```python
#7 )
def getemail(newdf):
    newerlist=[]
    for i,row in newdf.iterrows():
        newerlist.append(row['Email'].split('@',1)[0])
    newdf['Username']=newerlist
    return newdf
    
    
emaildf=pd.DataFrame()
emails=['username@domain.com','test@domain.com','username@gmail.com','ghost@invincible.com']
emaildf['Email']=emails
getemail(emaildf)
```

```python
#8) 
def filtereddf(newdf):
    filtereddf=pd.DataFrame()
    filtereddf['A']=newdf[newdf['A']>5]['A']
    filtereddf['B']=newdf[newdf['B']<10]['B']
    return filtereddf
    
mydict01={'A':[3,8,6,2,9],
 'B':[5,2,9,3,1],
 'C':[1,7,4,5,2]}
filterdf=pd.DataFrame(mydict01)
filtereddf(filterdf)
```

```python
#11) 
def adddaystodf(newdf):
    newdf['Weekday']=newdf['Date'].dt.day_name()
    return newdf
    
newdatelist=pd.date_range('2023-01-01','2023-01-28',freq='D').to_series()
adddaysdf=pd.DataFrame()
adddaysdf['Date']=newdatelist
adddaystodf(adddaysdf)
```

```python
#12) 
def gatedatesint(newdf):
    newerlist=newdf[(newdf['Date']>='2023-01-01') & (newdf['Date']<='2023-01-31')]
    return newerlist
    
newdatelist=pd.date_range('2023-01-01 00:00:00','2023-02-28 00:00:00', freq ='5H').to_series()
adddaysdf=pd.DataFrame()
adddaysdf['Date']=newdatelist
gatedatesint(adddaysdf)
```
