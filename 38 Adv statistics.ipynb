```python
#1) 

#assuming given sample size is 50
import scipy.stats as stats
import math

sample_mean = 50
sample_std = 5
confidence_level = 0.95
sample_size = 50

standard_error = sample_std / (math.sqrt(sample_size))

margin_of_error = stats.t.ppf((1 + confidence_level) / 2, sample_size - 1) * standard_error

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("confidence_interval is",confidence_interval)
```

```python
#7) 

import scipy.stats as stats

# Given data
total_sample_size = 1000
number_preferred_coke = 520
expected_proportion = 0.5  # Assuming no preference difference (50% prefer Coke)

sample_proportion = number_preferred_coke / total_sample_size

expected_frequency_coke = expected_proportion * total_sample_size
expected_frequency_pepsi = (1 - expected_proportion) * total_sample_size

observed_frequencies = [number_preferred_coke, total_sample_size - number_preferred_coke]
expected_frequencies = [expected_frequency_coke, expected_frequency_pepsi]

chi2, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-Value:", p_value)
```

```python
#8)

import scipy.stats as stats

observed_tails = 45
total_flips = 100

expected_tails = 0.5 * total_flips
expected_heads = 0.5 * total_flips

observed_frequencies = [observed_tails, total_flips - observed_tails]
expected_frequencies = [expected_tails, expected_heads]

chi2, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

alpha = 0.05

print("Chi-Square Statistic:", chi2)
print("P-Value:", p_value)

if p_value < alpha:
    print("The coin is biased towards tails.")
else:
    print("The coin is not biased towards tails.")
```

```python
#9)
import scipy.stats as stats

d = {'LungCancerYes': [60, 30], 'LungCancerNo': [140, 170]}

contingency_table = [
    d['LungCancerYes'],
    d['LungCancerNo']
]

chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

alpha = 0.05

print("Chi-Square Statistic:", chi2)
print("P-Value:", p_value)

if p_value < alpha:
    print("There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("There is no significant association between smoking status and lung cancer diagnosis.")

```

```python
#11

import scipy.stats as stats

sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70

t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=sample_size - 1))

alpha = 0.05

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

if p_value < alpha:
    print("The population mean is significantly different from 70.")
else:
    print("There is no significant evidence to conclude that the population mean is different from 70.")
```
