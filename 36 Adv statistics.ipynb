1) Estimation statistics is a branch of statistics that involves making inferences and estimating unknown population parameters based on sample data. It provides methods for estimating values of parameters, such as means, proportions, or variances, using information obtained from a sample.

Point Estimate:
A point estimate is a single value that is used to estimate an unknown population parameter. It is obtained by using a sample statistic, such as the sample mean or sample proportion, as an estimate of the corresponding population parameter.

Interval Estimate:
An interval estimate provides a range of values within which the unknown population parameter is likely to fall. It takes into account both the point estimate and the variability or uncertainty associated with the estimation process. The range is expressed as a confidence interval, which quantifies the level of confidence we have in capturing the true population parameter.

```python
#2)
import math

def popmean(samplemean,samplestd,samplesize):
    standarderror=samplestd/math.sqrt(samplesize)
    
    z_score=1.96
    
    marginoferror=z_score*standarderror
    
    lowerbound=samplemean-marginoferror
    upperbound=samplemean+marginoferror
    
    return samplemean,lowerbound,upperbound

popmean(60,50,10)
```

3) Hypothesis testing is a statistical procedure used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses, the null hypothesis (H₀) and the alternative hypothesis (H₁), and assessing the evidence from the sample to determine which hypothesis is more plausible.
 It helps ensure that findings are based on sound statistical reasoning, fostering reliable knowledge and informed decision-making.
 The importance of hypothesis testing lies in its ability to provide a systematic and rigorous approach to testing claims or hypotheses about population parameters. 

4) H0 (Null hypothesis)- the average weight of male college students is greater than the average weight of female college students.
   H1 (Alternate Hypothesis)- the average weight of male college students is not greater than the average weight of female college students.

```python
# 5) 

import numpy as np
import scipy.stats as stats

def hypothesis_test(sample1, sample2, alpha):
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    std1, std2 = np.std(sample1), np.std(sample2)
    
    se = np.sqrt((std1**2)/len(sample1) + (std2**2)/len(sample2))
    
    z_statistic = (mean1 - mean2) / se
    
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  
    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")
    
    print("z-statistic:", z_statistic)
    print("p-value:", p_value)

sample1 = [10, 12, 15, 14, 11]
sample2 = [8, 9, 12, 11, 10]
alpha = 0.05

hypothesis_test(sample1, sample2, alpha)
```

6) Null hypothesis is the base assumption where we dont have any relationship in variables.
Alternate hypothesis is opposite of null hypothesis where we assume there is a relationship in variables.

Eg- H0 (Null hypothesis)- the average weight of male college students is greater than the average weight of female college students. H1 (Alternate Hypothesis)- the average weight of male college students is not greater than the average weight of female college students.


7) Steps for hypothesis testing are - 
    1) Get the Null hypothesis
    2) Get the Alternate hypothesis
    3) Perform statistical tests
    4) decide whether to accept or reject the null hypothesis

8) P value is probability value that measures strength of evidence against the null hypothesis.
In hypothesis testing a small p-value indicates strong evidence against the null hypothesis, suggesting that the observed effect or difference is unlikely to be due to random chance alone.

```python
#10

import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    
    print("T-Statistic:", t_statistic)
    print("P-Value:", p_value)
    
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

sample1 = [10, 12, 15, 14, 11]
sample2 = [8, 9, 12, 11, 10]

two_sample_t_test(sample1, sample2)
```
