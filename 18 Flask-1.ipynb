5) The url_for() method is used for creating url in flask. 

from flask import Flask,url_for

app=Flask(__name__)

@app.route("/route/<variable>")
def Routeprint(variable):
    return f"{variable}\'s profile Company Name: ABC Corporation <br>Location: India <br>Contact Detail: 999-999-9999"  

with app.test_request_context():
  print(url_for('Routeprint', variable='Dhiraj'))

if __name__=="__main__":
    app.run(host="0.0.0.0")
